## LIBRERIAS

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

### REQUESTS + BEAUTIFULSOUP

```python
url = "http://finance.yahoo.com/quote/YPF"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Resumen de la noticia
soup.select("li", {'class': "js-stream-content"})
```

```python
def web_scrapper(soup):
    resultado = []

    items = soup.find_all('li', class_='js-stream-content')

    for item in items:
        titulo = item.find('h3').get_text()
        resumen = item.find('p').get_text()

        struct = {
            "Titulo": titulo,
            "Resumen": resumen
        }
        resultado.append(struct)

    return resultado
```

```python
data = web_scrapper(soup)
df = pd.DataFrame(data)
df
```
