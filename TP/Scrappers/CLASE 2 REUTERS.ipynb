```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.reuters.com/site-search/?query=YPF&offset=0"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}
response = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Utilizar un selector CSS m√°s general para encontrar las etiquetas li
reuters = soup.find_all("li", class_="search-result-item")
```

```python
# Utilizar un selector CSS m√°s general para encontrar las etiquetas li
reuters = soup.find_all("li.search-results_item_2oqiX", class_="search-result-item")
```

```python
news = []
```

```python
# Iterar sobre las etiquetas encontradas
for item in reuters:
    # Obtener el t√≠tulo de la noticia
    titulo = item.find("a", class_="text__text__1FZLe text__inherit-color__3208F text__inherit-font__1Y8w3 text__inherit-size__1DZJi link__link__3Ji6W link__underline_on_hover__2zGL4 title__link__1zUJP basic-card__title__3byru").text.strip()

    # Obtener la fecha de la noticia
    fecha = item.find("text__text__1FZLe text__medium-grey__3A_RT text__light__1nZjX text__ultra_small__37j9j date-line__time__1VR1r basic-card__time__316Dt").text.strip()

    # Obtener el resumen de la noticia, si est√° disponible
    resumen = item.find("p", class_="search-result-excerpt")
    resumen_text = resumen.text.strip() if resumen else ""

    # Agregar la noticia al listado
    news.append({"titulo": titulo, "fecha": fecha, "resumen": resumen_text})
```

```python
# Convertir el listado de noticias a un DataFrame de pandas
df = pd.DataFrame(news)
```
