```python
# Utilizar un selector CSS para encontrar las etiquetas h3 con la clase especificada
ft = soup.select("div.o-teaser__content")

news = []


# Iterar sobre las etiquetas h3 encontradas
for i in ft:
    # Obtener el texto del primer p√°rrafo (clase "byline")
    #print(zack)
    
    #h3_elements = soup.select("h3.truncated_text_single.listing_article_heading")

    titulo = i.find("a", class_="js-teaser-heading-link").text.strip()


    fecha = i.find("div", class_="o-teaser__timestamp").text.strip()
    
    # Obtener el texto del segundo p√°rrafo (clase "teaser")
    resumen = i.find("a", class_="js-teaser-standfirst-link").text.strip()
    
    
    temporal = {"titulo":titulo, "fecha":fecha, "resumen": resumen}

    news.append(temporal)

   
print(pd.DataFrame(news))

    
```

```python
import time
import random
import pandas as pd

def scrape_data(url):
    results = []

    page = 1
    while True:
        page_url = f"{url}{page}"
        
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
            }
        response = requests.get(page_url, headers=headers)
        
        if response.status_code != 200:
            print(f"Error 200 en p√°gina {page}.")
            break

        soup = BeautifulSoup(response.content, 'html.parser')

        items = soup.select("div.o-teaser__content")

        if len(items)==0 or items is None:
            print(f"La p√°gina {page} no existe o no contiente datos.")
            break

        for item in items:
            date_span = item.find("div", class_="o-teaser__timestamp").text.strip()
            title = item.find("a", class_="js-teaser-heading-link").text.strip()

            result = {
                "Fecha": date_span,
                "Titulo": title,
                "Fuente": "Financial Times"
            }
            results.append(result)

        print(f"P√°gina {page} procesada")
        page += 1
        time.sleep(random.randint(0, 5))

    return results
```

```python
url_base = "https://www.ft.com/search?q=YPF&page="

resultados = scrape_data(url_base)
df_resultados = pd.DataFrame(resultados)
```

### Formato

```python
from datetime import datetime

df_resultados_fecha = df_resultados

df_resultados_fecha['Fecha'] = df_resultados['Fecha'].apply(lambda fecha_str: datetime.strptime(fecha_str, "%B %d, %Y").strftime("%d/%m/%Y"))
```
